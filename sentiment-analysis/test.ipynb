```python
import pandas as pd
import matplotlib.pyplot as plt

# create a sample dataframe
df = pd.DataFrame({
    'year': [2018, 2018, 2018, 2018, 2019, 2019, 2019, 2019, 2020, 2020, 2020, 2020, 2021, 2021, 2021, 2021, 2022, 2022, 2022, 2022, 2023, 2023, 2023, 2023],
    'quarter': [1, 2, 3, 4]*6,
    'measurement': [10, 12, 15, 13, 11, 16, 17, 14, 9, 8, 10, 11, 13, 12, 15, 18, 20, 22, 23, 25, 30, 32, 33, 35]
})

# create a new column with concatenated year and quarter
df['year_quarter'] = df['year'].astype(str) + ' Q' + df['quarter'].astype(str)

# pivot the dataframe to get each measurement as a separate column
pivoted_df = df.pivot(index='year_quarter', columns='year', values='measurement')

# plot the pivoted dataframe
pivoted_df.plot(marker='o')

# add plot labels
plt.xlabel('Year Quarter')
plt.ylabel('Measurement')
plt.title('Time Series of Measurements by Quarter')
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({'colors': ['red', 'green', 'blue', 'red', 'green', 'green']})

# use factorize to encode the 'colors' column
df['colors_encoded'], unique_colors = pd.factorize(df['colors'])

# print the encoded DataFrame and the unique colors
print(df)
print(unique_colors)
```
